what is linear programming
-  a maximization or minimization problem based on linear systems of equations/inequality
   
what are the conditions for LP to be used?
- when objective functions and constraints are linear

What is a feasible region?
- a set of solutions where they satisties linear constraints

How many intersection can multiple linear constraints have?
- exponential  2^n runtime worst case

what are slack and excess variables?
- slack is a variable that indicate how much the value is far away from reaching the constraints while within the feasible region

what is a basic and non basic variable?
- non basic variables are the variable we set to zero, while we solve for basic variable satisfying the constraints

Describe the simplex method? ( ratio test, object function pivot)
- for max problem, choose the most negative objective function coefficients(non basic), and do row reduction for row with the at least positive ratio test, and, and the basic variable in that row will left and swap with the non-basic column
- iterate this method until theres no more negative objective funtion coefficients

what are the condition for unboundness?
- ratio tests cant be performed

what are the condition for multiple optimal solution?
- when iterating the method result in same objective value (z)

what is a degeneracy
